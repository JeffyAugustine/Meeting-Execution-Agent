```python
# Setup and imports
import sys
import os
sys.path.append('..')

from dotenv import load_dotenv
load_dotenv()

import src.ingest as ingest
from src.understand import analyze_meeting, print_analysis_results
from src.validate import validate_tasks, deduplicate_tasks
from src.planner import plan_tasks
from src.action import generate_followup_email, export_to_csv, export_to_json
from src.dashboard import generate_task_dashboard, display_task_summary
from src.evaluate import run_comprehensive_evaluation, print_evaluation_summary

print("âœ… All modules loaded successfully!")
```

#  1. Choose a Meeting to Analyze

```python
# List available meetings
meeting_files = [f for f in os.listdir('../data/sample_transcripts') if f.endswith('.txt')]
print("ğŸ“ Available meetings:")
for i, file in enumerate(meeting_files, 1):
    print(f"   {i}. {file}")
```

```python
# Use meeting_01 for demo
selected_meeting = 'meeting_01.txt'
print(f"\nğŸ¯ Selected: {selected_meeting}")
```

#  2. Process Meeting Transcript

```python
print("ğŸ”„ Processing meeting transcript...")
transcript_text = ingest.process_transcript(f'../data/sample_transcripts/{selected_meeting}')

print(f"ğŸ“Š Transcript stats:")
print(f"   â€¢ Characters: {len(transcript_text)}")
print(f"   â€¢ Words: {len(transcript_text.split())}")

print("\nğŸ“ Preview:")
print("=" * 50)
print(transcript_text[:600] + "..." if len(transcript_text) > 600 else transcript_text)
print("=" * 50)
```

#  3. AI-Powered Task Extraction

```python
print("ğŸ¤– Analyzing with Gemini AI...")
analysis_results = analyze_meeting(transcript_text)
```

```python
print_analysis_results(analysis_results)
```

#  4. Validate & Plan Execution

```python
print("âœ… Validating and planning tasks...")
tasks = analysis_results.get('tasks', [])
participants = analysis_results.get('participants', [])

# Validate tasks
validated_tasks = validate_tasks(tasks, participants)
valid_owners = sum(1 for task in validated_tasks if task.get('owner_valid', False))
print(f"   â€¢ Valid owners: {valid_owners}/{len(validated_tasks)}")

# Deduplicate
deduplicated_tasks = deduplicate_tasks(validated_tasks)
print(f"   â€¢ After deduplication: {len(deduplicated_tasks)} tasks")

# Add execution plans
planned_tasks = plan_tasks(deduplicated_tasks)
print(f"   â€¢ Execution plans generated")

# Show task summary
display_task_summary(planned_tasks)
```

#  5. Generate Outputs

```python
print("ğŸ“¤ Generating outputs...")

# Generate follow-up email
email_subject, email_body = generate_followup_email(
    analysis_results.get('meeting_summary', ''), 
    planned_tasks, 
    participants
)

print("ğŸ“§ FOLLOW-UP EMAIL DRAFT:")
print("=" * 50)
print(f"Subject: {email_subject}")
print(email_body)
print("=" * 50)

# Export files
export_to_csv(planned_tasks, '../assets/demo_tasks.csv')
export_to_json(planned_tasks, '../assets/demo_tasks.json')
print(f"\nğŸ’¾ Exported: demo_tasks.csv, demo_tasks.json")

# Generate dashboard
html_dashboard = generate_task_dashboard(planned_tasks)
with open('../assets/demo_dashboard.html', 'w', encoding='utf-8') as f:
    f.write(html_dashboard)
print(f"ğŸŒ Dashboard: demo_dashboard.html")
```
